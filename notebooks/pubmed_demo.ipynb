```python
# Required imports and setup - DO NOT MODIFY
from pprint import pprint
import requests

BASE_URL = "http://analytics-staging.ga4gh.org:8000"

def print_response(response):
    """
    Helper function to display API responses in a readable format.

    What it does:
    - Shows if the request was successful (status code 200)
    - Displays the data in an organized way
    - Shows error messages if something went wrong

    DO NOT MODIFY: This function is used throughout the notebook
    """
    print(f"Status Code: {response.status_code}")
    if response.status_code == 200:
        try:
            data = response.json()
            print("Response:")
            pprint(data)
        except:
            print("Response (text):", response.text)
    else:
        print("Error:", response.text)
```

```python
from collections import Counter
from datetime import datetime

import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

**First, fetch article data from the API for a given keyword.**

```python
def get_articles_data(keyword):
    """
    Fetches article data from the API for a given keyword.

    What it does:
    - Sends a request to our analytics dashboard API
    - Returns a list of articles if successful
    - Returns an empty list if there's an error
    - Prints error messages to help with troubleshooting

    Parameters:
    - keyword: The search term (e.g., "GA4GH")

    Returns:
    - List of article dictionaries with all the metadata

    DO NOT MODIFY: This function is used throughout the analytics
    """
    url = f"{BASE_URL}/pubmed/articles/{keyword}"
    response = requests.get(url)

    if response.status_code == 200:
        return response.json()
    print(f"Error fetching data for {keyword}: {response.status_code}")
    return []
```

### 5.2 Publication Timeline Analysis (Do not change anything here)

```python
def parse_publication_dates(articles):
    """
    Extract and parse publication dates from articles with consistent timezone handling.

    What it does:
    - Takes publication dates from articles
    - Converts them to a standard format
    - Handles timezone information consistently
    - Returns a list of datetime objects for analysis

    DO NOT MODIFY: This handles complex date parsing and timezone issues
    """
    dates = []
    for article in articles:
        pub_date = article.get("publish_date")
        if pub_date:
            try:
                # Parse ISO format date and handle timezone consistently
                if "Z" in pub_date:
                    date_obj = datetime.fromisoformat(pub_date.replace("Z", "+00:00"))
                elif "+" in pub_date or pub_date.count("-") > 2:
                    date_obj = datetime.fromisoformat(pub_date)
                else:
                    # Parse as naive datetime and treat as UTC
                    date_obj = datetime.fromisoformat(pub_date)
                    # Remove timezone info to make all dates naive for consistent comparison
                    if date_obj.tzinfo is not None:
                        date_obj = date_obj.replace(tzinfo=None)

                dates.append(date_obj)
            except Exception as e:
                # Skip problematic dates rather than failing
                continue
    return dates

print("Creating Publication Timeline Analysis for all keywords")
print()

# Create subplot figure with dynamic number of timelines
from plotly.subplots import make_subplots

# Analyze publication timeline for each keyword
timelines_created = 0
fig = make_subplots(
    rows=len(keywords), cols=1,
    subplot_titles=[f"{keyword} Publications by Year" for keyword in keywords],
    vertical_spacing=0.15
)

for i, keyword in enumerate(keywords, 1):
    articles = all_articles_data.get(keyword, [])
    dates = parse_publication_dates(articles)

    if dates:
        # Convert to DataFrame for easier manipulation
        df = pd.DataFrame({"date": dates})
        df["year"] = df["date"].dt.year

        # Publications by year
        yearly_counts = df["year"].value_counts().sort_index()

        # Add yearly bar chart for this keyword
        fig.add_trace(
            go.Bar(
                x=yearly_counts.index,
                y=yearly_counts.values,
                name=f"{keyword} Publications",
                marker_color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'][i-1 % 5],
                text=yearly_counts.values,
                textposition='outside',
                showlegend=False
            ),
            row=i, col=1
        )

        # Update axis labels for each subplot
        fig.update_xaxes(title_text="Year", row=i, col=1)
        fig.update_yaxes(title_text="Number of Publications", row=i, col=1)

        timelines_created += 1

        print(f"{keyword}:")
        print(f"   Total publications with dates: {len(dates)}")
        if dates:
            print(f"   Date range: {min(dates).strftime('%Y-%m-%d')} to {max(dates).strftime('%Y-%m-%d')}")
        print()
    else:
        print(f" {keyword}: No publication dates found for timeline analysis")
        print()

if timelines_created > 0:
    # Update overall layout with more space between graphs
    fig.update_layout(
        height=max(500, timelines_created * 400),
        title_text="Publication Timeline Analysis - All Keywords",
        title_x=0.5,
        title_font_size=16
    )

    fig.show()

    print(f"Timeline Analysis Complete")
    print(f"   Created timelines for {timelines_created} out of {len(keywords)} keywords")
    print(f"   Keywords with data: {[k for k in keywords if parse_publication_dates(all_articles_data.get(k, []))]}")
else:
    print(" No timeline data available for any keywords")
    print("This might mean:")
    print("  - The articles don't have publication dates")
    print("  - The date format is not recognized")
    print("  - No articles were found for the keywords")
```

### 5.3 Top Journals Analysis
✨ top_journals_to_show can be modified here to show less/more journals in the result. This is to ask the question: what are the top x journals that have had an article published in for a specific keyword?

```python
def analyze_journals(articles, keyword, top_n=10):
    """
    Analyzes journal distribution for a specific keyword's articles.

    What it does:
    - Counts how many articles appear in each journal
    - Identifies the top publishing venues
    - Creates a horizontal bar chart visualization
    - Shows which journals are most active in this topic

    DO NOT MODIFY ANYTHING HERE
    """
    journals = [article.get("journal", "Unknown") for article in articles if article.get("journal")]

    if not journals:
        print(f"No journal data found for {keyword}")
        return

    journal_counts = Counter(journals)
    top_journals = journal_counts.most_common(top_n)

    # Create horizontal bar chart with Plotly
    journals_list, counts_list = zip(*top_journals, strict=False)

    # Truncate long journal names for display
    display_journals = [journal[:50] + "" if len(journal) > 50 else journal for journal in journals_list]

    fig = go.Figure(go.Bar(
        x=counts_list,
        y=display_journals,
        orientation='h',
        marker_color='lightcoral',
        text=counts_list,
        textposition='outside'
    ))

    fig.update_layout(
        title=f"Top {top_n} Journals for {keyword} Articles",
        xaxis_title="Number of Publications",
        yaxis_title="Journals",
        height=max(400, len(top_journals) * 40),
        margin=dict(l=200)  # Left margin for journal names
    )

    fig.show()

    print(f"\nTop {top_n} journals for {keyword}:")
    for i, (journal, count) in enumerate(top_journals, 1):
        print(f"  {i}. {journal}: {count} articles")

print("Creating Journal Analysis for all keywords")
print("This shows which journals publish articles for each keyword.")
print()

# ✨ CUSTOMIZABLE: Change this number to show more/fewer top journals
top_journals_to_show = 8  # Shows top 8 journals for each keyword

journals_analyzed = 0

# Analyze journals for each keyword in your list
for keyword in keywords:
    articles = all_articles_data.get(keyword, [])

    if articles:
        print(f"Analyzing journals for: {keyword}")
        print("-" * 60)

        analyze_journals(articles, keyword, top_n=top_journals_to_show)
        journals_analyzed += 1

        print("\n" + "="*80 + "\n")
    else:
        print(f"No articles found for {keyword} - skipping journal analysis")
        print()

print(f"\nJournal Analysis Complete")
print(f"Analyzed journals for {journals_analyzed} out of {len(keywords)} keywords")
if journals_analyzed > 0:
    print("Use this data to:")
    print("   • Identify the leading journals in your keyword")
    print("   • Track which journals are most active in different topics")
    print("   • Discover interdisciplinary publication patterns")
```

### 5.4 Author Analysis
✨ top_authors_to_show can be modified here to show less/more journals in the result. This is to ask the question: who are the top x authors who have published articles for a specific keyword?

```python
def analyze_authors(articles, keyword):
    """
    Analyzes author data from articles for a specific keyword.

    What it does:
    - Counts how many articles each author has published
    - Identifies the most published authors
    - Analyzes collaboration patterns (authors per article)
    - Creates visualizations showing author activity

    This helps you understand:
    - Who are the leading authors in this area

    DO NOT MODIFY: Complex analysis and visualization logic
    """
    all_authors = []
    author_article_count = Counter()

    for article in articles:
        authors = article.get("authors", [])
        for author in authors:
            author_name = author.get("name", "Unknown")
            all_authors.append(author_name)
            author_article_count[author_name] += 1

    if not all_authors:
        print(f"No author data found for {keyword}")
        return

    # Get top authors

    # ✨ CUSTOMIZABLE: Change this number to show more/fewer authors
    top_authors_to_show = 5
    top_authors = author_article_count.most_common(top_authors_to_show)

    # Create individual charts instead of subplots
    if top_authors:
        # Chart 1: Top authors horizontal bar chart
        authors_list, counts_list = zip(*top_authors, strict=False)
        display_authors = [name[:30] + "" if len(name) > 30 else name for name in authors_list]

        fig1 = go.Figure()
        fig1.add_trace(
            go.Bar(
                x=counts_list,
                y=display_authors,
                orientation='h',
                marker_color='lightgreen',
                text=counts_list,
                textposition='outside',
                name="Author Articles"
            )
        )

        fig1.update_layout(
            title=f"Top {top_authors_to_show} Authors for {keyword}",
            xaxis_title="Number of Publications",
            yaxis_title="Authors",
            height=400,
            margin=dict(l=200)
        )

        fig1.show()

    # Author statistics
    total_authors = len(set(all_authors))
    avg_authors_per_article = len(all_authors) / len(articles) if articles else 0

    print(f"\nAuthor Statistics for {keyword}:")
    print(f"   Total unique authors: {total_authors}")
    print(f"   Average authors per article: {avg_authors_per_article:.2f}")
    print(f"   Articles analyzed: {len(articles)}")

    if top_authors:
        print(f"\nTop {top_authors_to_show} most published authors:")
        for i, (author, count) in enumerate(top_authors[:5], 1):
            print(f"   {i}. {author}: {count} articles")

    return {
        'total_authors': total_authors,
        'avg_authors_per_article': avg_authors_per_article,
        'top_authors': top_authors,
        'articles_count': len(articles)
    }

print("Creating Author Analysis for all keywords")
print()

authors_analyzed = 0
all_keyword_stats = {}

# Analyze authors for each keyword in your list
for keyword in keywords:
    articles = all_articles_data.get(keyword, [])

    if articles:
        print(f"Analyzing authors for: {keyword}")
        print("-" * 70)

        keyword_stats = analyze_authors(articles, keyword)
        if keyword_stats:
            all_keyword_stats[keyword] = keyword_stats
            authors_analyzed += 1

        print("\n" + "="*80 + "\n")
    else:
        print(f"No articles found for {keyword} - skipping author analysis")
        print()

print(f"\nAuthor Analysis Complete")
print(f"Analyzed authors for {authors_analyzed} out of {len(keywords)} keywords")
if authors_analyzed > 0:
    print("Use this data to:")
    print("   • Identify leading authors in your field")
    print("   • Find potential collaboration partners")
    print("   • Understand author community structure")
    print("   • Track collaboration patterns and team sizes")
```

### 5.5 Comprehensive Dashboard Summary (Do not change anything here)

```python
def create_analytics_dashboard():
    """
    Creates a comprehensive analytics dashboard using data from all keywords in the keywords list.

    What this function does:
    - Analyzes articles from all keywords defined in the keywords list (cell-15)
    - Creates interactive visualizations showing patterns across all keyword areas
    - Generates summary statistics and comparison charts
    - Provides publication quality metrics and author community insights
    - Adapts automatically to any keywords you've configured

    The dashboard includes:
    - Article count comparisons across all keywords
    - Recent publication trends (2020+)
    - Publication status distribution for the first keyword
    - Top journals across all keyword topics
    - Comprehensive summary statistics table with key metrics

    This dashboard is fully generalized and will work with any keywords you configure.
    """
    dashboard_data = {}

    # Process all keywords from the keywords list (defined in cell-15)
    for keyword, articles in all_articles_data.items():
        if articles:
            total_articles = len(articles)

            # Analyze publication status distribution
            statuses = [article.get("status", "Unknown") for article in articles]
            status_counts = Counter(statuses)

            # Analyze journal distribution
            journals = [article.get("journal", "Unknown") for article in articles if article.get("journal")]
            journal_counts = Counter(journals)

            # Analyze author collaboration patterns
            unique_authors = set()
            for article in articles:
                for author in article.get("authors", []):
                    author_name = author.get("name", "Unknown")
                    if author_name == "Unknown":
                        unique_authors.add(author_name)
            total_unique_authors = len(unique_authors)

            # Analyze publication timeline
            dates = parse_publication_dates(articles)

            # Store comprehensive analytics for this keyword
            dashboard_data[keyword] = {
                "total_articles": total_articles,
                "status_counts": status_counts,
                "top_journals": journal_counts.most_common(3),
                "total_authors": total_unique_authors,
                "date_range": (min(dates), max(dates)) if dates else None,
                "recent_articles": len([d for d in dates if d.year >= 2020]) if dates else 0
            }

    # Create comprehensive dashboard with Plotly subplots
    from plotly.subplots import make_subplots

    fig = make_subplots(
        rows=3, cols=2,
        subplot_titles=(
            "Total Articles by Keyword",
            "Recent Articles (2020+)",
            f"{list(dashboard_data.keys())[0]} Article Status Distribution" if dashboard_data else "Article Status Distribution",
            "Top Journals (All Keywords)",
            "Summary Statistics",
            ""
        ),
        specs=[
            [{"type": "bar"}, {"type": "bar"}],
            [{"type": "pie"}, {"type": "bar"}],
            [{"type": "table", "colspan": 2}, None]
        ],
        vertical_spacing=0.15,
        horizontal_spacing=0.25  # Increased from 0.1 to 0.25 for more space between side-by-side graphs
    )

    # 1. Overall article counts across all keywords
    keywords_list = list(dashboard_data.keys())
    article_counts = [dashboard_data[k]["total_articles"] for k in keywords_list]

    # Use different colors for each keyword (cycles through colors if more keywords than colors)
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']
    bar_colors = [colors[i % len(colors)] for i in range(len(keywords_list))]

    fig.add_trace(
        go.Bar(
            x=keywords_list,
            y=article_counts,
            name="Total Articles",
            marker_color=bar_colors,
            text=article_counts,
            textposition='outside'
        ),
        row=1, col=1
    )

    # 2. Recent articles (2020+) across all keywords
    recent_counts = [dashboard_data[k]["recent_articles"] for k in keywords_list]

    fig.add_trace(
        go.Bar(
            x=keywords_list,
            y=recent_counts,
            name="Recent Articles",
            marker_color='lightgreen',
            text=recent_counts,
            textposition='outside'
        ),
        row=1, col=2
    )

    # 3. Status distribution for first keyword (if available)
    if keywords_list:
        first_keyword = keywords_list[0]
        status_data = dashboard_data[first_keyword]["status_counts"]

        fig.add_trace(
            go.Pie(
                labels=list(status_data.keys()),
                values=list(status_data.values()),
                name="Status Distribution"
            ),
            row=2, col=1
        )

    # 4. Top journals across all keywords
    all_journals = Counter()
    for data in dashboard_data.values():
        for journal, count in data["top_journals"]:
            all_journals[journal] += count

    top_all_journals = all_journals.most_common(5)
    if top_all_journals:
        journals, counts = zip(*top_all_journals, strict=False)
        display_journals = [j[:25] + "" if len(j) > 25 else j for j in journals]

        fig.add_trace(
            go.Bar(
                x=counts,
                y=display_journals,
                orientation='h',
                name="Top Journals",
                marker_color='coral',
                text=counts,
                textposition='outside'
            ),
            row=2, col=2
        )

    # 5. Summary Statistics - Add comprehensive metrics as a table
    if dashboard_data:
        # Calculate overall statistics
        total_articles = sum(data["total_articles"] for data in dashboard_data.values())
        total_authors = sum(data["total_authors"] for data in dashboard_data.values())
        total_recent = sum(data["recent_articles"] for data in dashboard_data.values())

        # Calculate averages and percentages
        avg_articles_per_keyword = total_articles / len(dashboard_data) if len(dashboard_data) > 0 else 0
        avg_authors_per_keyword = total_authors / len(dashboard_data) if len(dashboard_data) > 0 else 0
        recent_percentage = (total_recent / total_articles * 100) if total_articles > 0 else 0

        # Create summary statistics table
        summary_stats = [
            ["Keywords Analyzed", len(dashboard_data)],
            ["Total Articles", total_articles],
            ["Total Unique Authors", total_authors],
            ["Recent Articles (2020+)", f"{total_recent} ({recent_percentage:.1f}%)"],
            ["Avg Articles per Keyword", f"{avg_articles_per_keyword:.1f}"],
            ["Avg Authors per Keyword", f"{avg_authors_per_keyword:.1f}"]
        ]

        # Add top performing keyword if multiple keywords exist - show full keyword names
        if len(dashboard_data) > 1:
            most_articles = max(dashboard_data.items(), key=lambda x: x[1]["total_articles"])
            most_collaborative = max(dashboard_data.items(), key=lambda x: x[1]["total_authors"])

            summary_stats.extend([
                ["Most Articles", f"{most_articles[0]} ({most_articles[1]['total_articles']})"],
                ["Most Collaborative", f"{most_collaborative[0]} ({most_collaborative[1]['total_authors']} authors)"]
            ])

        fig.add_trace(
            go.Table(
                header=dict(
                    values=["Metric", "Value"],
                    fill_color='#3498DB',
                    font=dict(color='white', size=12),
                    align="left",
                    height=35
                ),
                cells=dict(
                    values=[
                        [stat[0] for stat in summary_stats],
                        [stat[1] for stat in summary_stats]
                    ],
                    fill_color=[['#ECF0F1', '#D5DBDB'] * (len(summary_stats) // 2 + 1)][:len(summary_stats)],
                    font=dict(size=11),
                    align="left",
                    height=30
                )
            ),
            row=3, col=1
        )

    # Update layout
    fig.update_layout(
        height=1200,
        showlegend=False,
        title_text="PubMed Analytics Dashboard - All Keywords",
        title_x=0.5,
        title_font_size=18
    )

    # Update axis labels
    fig.update_xaxes(title_text="Keywords", row=1, col=1)
    fig.update_yaxes(title_text="Number of Articles", row=1, col=1)
    fig.update_xaxes(title_text="Keywords", row=1, col=2)
    fig.update_yaxes(title_text="Number of Articles", row=1, col=2)
    fig.update_xaxes(title_text="Number of Articles", row=2, col=2)
    fig.update_yaxes(title_text="Journals", row=2, col=2)

    fig.show()

    # Create summary table with all keywords
    summary_df = pd.DataFrame([
        {
            "Keyword": keyword,
            "Total Articles": data["total_articles"],
            "Unique Authors": data["total_authors"],
            "Recent (2020+)": data["recent_articles"],
            "Top Journal": data["top_journals"][0][0][:30] + "" if data["top_journals"] and len(data["top_journals"][0][0]) > 30 else data["top_journals"][0][0] if data["top_journals"] else "N/A"
        }
        for keyword, data in dashboard_data.items()
    ])

    # Display summary table
    from plotly.graph_objects import Figure, Table

    table_fig = go.Figure(data=[go.Table(
        header=dict(
            values=list(summary_df.columns),
            fill_color='#4CAF50',
            font=dict(color='white', size=12),
            align="center"
        ),
        cells=dict(
            values=[summary_df[col] for col in summary_df.columns],
            fill_color='lightgrey',
            align="center",
            font=dict(size=11)
        )
    )])

    table_fig.update_layout(
        title="Analytics Summary Table - All Keywords",
        height=max(300, len(summary_df) * 50 + 100)  # Dynamic height based on number of keywords
    )

    table_fig.show()

    # Print comprehensive summary
    print("=== COMPREHENSIVE ANALYTICS SUMMARY ===")
    total_all_articles = sum(data["total_articles"] for data in dashboard_data.values())
    total_all_authors = sum(data["total_authors"] for data in dashboard_data.values())

    print(f"Total articles across all keywords: {total_all_articles}")
    print(f"Total unique authors: {total_all_authors}")
    print(f"Keywords analyzed: {len(dashboard_data)}")
    print(f"Keywords list: {list(dashboard_data.keys())}")

    # Show detailed breakdown for each keyword
    for keyword, data in dashboard_data.items():
        print(f"\n{keyword}:")
        print(f"  Articles: {data['total_articles']}")
        print(f"  Unique authors: {data['total_authors']}")
        print(f"  Recent (2020+): {data['recent_articles']}")
        if data["date_range"]:
            start_date, end_date = data["date_range"]
            print(f"  Date range: {start_date.strftime('%Y-%m-%d')} to {end_date.strftime('%Y-%m-%d')}")

        # Show top status for this keyword
        if data["status_counts"]:
            top_status = data["status_counts"].most_common(1)[0]
            print(f"  Primary status: {top_status[0]} ({top_status[1]} articles)")

        # Show top journal for this keyword
        if data["top_journals"]:
            print(f"  Top journal: {data['top_journals'][0][0]} ({data['top_journals'][0][1]} articles)")

# Create the generalized dashboard
print("Creating Comprehensive Analytics Dashboard")
print("This dashboard analyzes all keywords from your configured list.")
print()

create_analytics_dashboard()

print("\nComprehensive Dashboard Complete")
print("This dashboard is fully generalized and will work with any keywords you configure.")
print("To analyze different topics:")
print("   • Modify the keywords list in cell-15")
print("   • Re-run from cell-15 onwards to fetch new data")
```

```python
def create_rews_analytics_table():
    """
    Creates a comprehensive analytics summary table for REWS keywords.

    What this function does:
    - Analyzes each keyword's articles for patterns and statistics
    - Counts publications, authors, journals, and publication status
    - Creates an interactive table with all the key metrics
    - Provides summary statistics and insights

    The table includes:
    - Article counts and publication status (published vs preprint)
    - Journal diversity and top publishing venues
    - Author collaboration patterns
    - Publication date ranges and recent activity

    DO NOT MODIFY
    """

    rews_analytics = []

    for keyword, articles in rews_articles_data.items():
        if not articles:
            continue

        # Basic metrics - count total articles
        total_articles = len(articles)

        # Status analysis - count published vs preprint articles
        statuses = [article.get("status", "Unknown") for article in articles]
        status_counts = Counter(statuses)
        published_count = status_counts.get("Published", 0)
        preprint_count = status_counts.get("Preprint", 0)

        # Journal analysis - diversity and top venues
        journals = [article.get("journal", "Unknown") for article in articles if article.get("journal")]
        unique_journals = len(set(journals))
        top_journal = Counter(journals).most_common(1)[0] if journals else ("N/A", 0)

        # Author analysis - collaboration patterns
        all_authors = []
        for article in articles:
            for author in article.get("authors", []):
                author_name = author.get("name", "Unknown")
                if author_name == "Unknown":
                    all_authors.append(author_name)

        unique_authors = len(set(all_authors))
        avg_authors_per_article = len(all_authors) / total_articles if total_articles > 0 else 0

        # Date analysis - publication timeline
        dates = parse_publication_dates(articles)
        if dates:
            date_range = f"{min(dates).strftime('%Y')} - {max(dates).strftime('%Y')}"
            recent_articles = len([d for d in dates if d.year >= 2020])
        else:
            date_range = "N/A"
            recent_articles = 0


        # Create analytics entry for the table
        analytics_entry = {
            "Keyword": get_clean_keyword_name(keyword),
            "Total Articles": total_articles,
            "Published": published_count,
            "Preprint": preprint_count,
            "Unique Journals": unique_journals,
            "Top Journal": top_journal[0][:40] + "" if len(top_journal[0]) > 40 else top_journal[0],
            "Top Journal Count": top_journal[1],
            "Unique Authors": unique_authors,
            "Avg Authors/Article": round(avg_authors_per_article, 1),
            "Date Range": date_range,
            "Recent (2020+)": recent_articles,
        }

        rews_analytics.append(analytics_entry)

    # Convert to DataFrame for easier handling
    df = pd.DataFrame(rews_analytics)

    # Create interactive table with professional styling
    fig = go.Figure(data=[go.Table(
        header=dict(
            values=list(df.columns),
            fill_color='#2E86C1',  # Professional blue header
            font=dict(color='white', size=11),
            align="center",
            height=40
        ),
        cells=dict(
            values=[df[col] for col in df.columns],
            fill_color=[['#E8F6F3', '#FADBD8'] * (len(df) // 2 + 1)][:len(df)],  # Alternating row colors
            align="center",
            font=dict(size=10),
            height=35
        )
    )])

    # Configure the table layout and title
    fig.update_layout(
        title=dict(
            text="<b>REWS Analytics Summary Table</b><br><sub>Regulatory & Ethics Work Stream - GA4GH Keywords Analysis</sub>",
            x=0.5,
            font=dict(size=16)
        ),
        height=400,
        margin=dict(l=20, r=20, t=80, b=20)
    )

    # Display the interactive table
    fig.show()

    # Print detailed analysis summary
    print("REWS ANALYTICS SUMMARY")
    print("=" * 50)

    # Calculate overall metrics
    total_rews_articles = sum(entry["Total Articles"] for entry in rews_analytics)
    total_rews_authors = sum(entry["Unique Authors"] for entry in rews_analytics)
    total_rews_journals = sum(entry["Unique Journals"] for entry in rews_analytics)

    print(f"Overall REWS Metrics:")
    print(f"   Total Articles: {total_rews_articles}")
    print(f"   Total Unique Authors: {total_rews_authors}")
    print(f"   Total Unique Journals: {total_rews_journals}")
    print(f"   Keywords Analyzed: {len(rews_analytics)}")

    # Show top performing keywords
    print(f"\nTop Performing REWS Keywords:")
    sorted_by_articles = sorted(rews_analytics, key=lambda x: x["Total Articles"], reverse=True)
    for i, entry in enumerate(sorted_by_articles[:3], 1):
        print(f"   {i}. {entry['Keyword']}: {entry['Total Articles']} articles")

    # Publication quality analysis
    print(f"\nPublication Quality:")
    total_published = sum(entry["Published"] for entry in rews_analytics)
    total_preprint = sum(entry["Preprint"] for entry in rews_analytics)
    if total_rews_articles > 0:
        print(f"   Published: {total_published} ({total_published/total_rews_articles*100:.1f}%)")
        print(f"   Preprint: {total_preprint} ({total_preprint/total_rews_articles*100:.1f}%)")

    return df

def get_clean_keyword_name(keyword):
    """
    Removes the long 'Global Alliance for Genomics and Health' prefix for cleaner display.

    What it does:
    - Takes a long keyword name
    - Removes the standard GA4GH prefix
    - Returns a shorter, more readable version

    Example:
    Input: "Global Alliance for Genomics and Health Your DNA, Your Say"
    Output: "Your DNA, Your Say"

    DO NOT MODIFY: Used for consistent naming throughout the notebook
    """
    prefix = "Global Alliance for Genomics and Health "
    if keyword.startswith(prefix):
        return keyword[len(prefix):]
    return keyword

def get_short_name(keyword):
    """
    Creates very short display names for charts and limited space displays.

    What it does:
    - Maps long keyword names to short, memorable versions
    - Used in charts where space is limited
    - Provides fallback for unknown keywords

    Just change the text after the = sign in each condition
    """
    if "Your DNA, Your Say" in keyword:
        return "Your DNA, Your Say"
    elif "Regulatory & Ethics" in keyword:
        return "Regulatory & Ethics"
    elif "Diversity in Datasets" in keyword:
        return "Diversity in Datasets"
    elif "Framework for responsible sharing" in keyword:
        return "Responsible Sharing Framework"
    elif "Health Policy on Clinically Actionable" in keyword:
        return "Clinically Actionable Results"
    else:
        clean_name = get_clean_keyword_name(keyword)
        return clean_name[:30] + "" if len(clean_name) > 30 else clean_name

print("Creating REWS Analytics Summary Table")
print()

rews_analytics_df = create_rews_analytics_table()

print("\nREWS Analytics Table Complete")
```

## Detailed Article Tables

**What this section does:** Creates detailed tables showing all the information for each article found in the searches.

**You'll see:**
- Individual tables for each keyword
- One comprehensive table with all articles combined
- Complete article details: titles, authors, journals, abstracts, publication dates

```python
def create_detailed_articles_tables():
    """
    Creates detailed tables for each REWS keyword showing all article fields.

    What this function does:
    - Takes each keyword and creates a separate table for its articles
    - Shows: title, journal, authors, publication date, status, abstract
    - Truncates long text for readability but preserves full information
    - Provides summary statistics for each keyword
    - Creates interactive tables you can scroll through

    For each article, you'll see:
    - Full title (truncated if very long)
    - Journal name where it was published
    - First 3 authors (with count if more)
    - Publication date (formatted as YYYY-MM-DD)
    - Publication status (published, preprint, etc.)
    - Abstract preview (first 100 characters)

    DO NOT MODIFY: Complex table creation logic
    """

    for keyword, articles in rews_articles_data.items():
        if not articles:
            print(f"No articles found for '{get_clean_keyword_name(keyword)}'")
            continue

        clean_keyword = get_clean_keyword_name(keyword)
        print(f"\nDETAILED ARTICLES TABLE: {clean_keyword}")
        print("=" * 80)

        # Prepare data for the table
        table_data = []

        for i, article in enumerate(articles, 1):
            # Extract all available fields with safe handling of missing data
            title = article.get("title") or "N/A"
            abstract = article.get("abstract") or "N/A"
            journal = article.get("journal") or "N/A"
            publish_date = article.get("publish_date") or "N/A"
            status = article.get("status") or "N/A"

            # Format authors - show first 3 with count if more exist
            authors_list = article.get("authors", [])
            if authors_list:
                authors_names = [author.get("name", "Unknown") for author in authors_list]
                authors_str = "; ".join(authors_names[:3])  # 🎛️ Change 3 to show more/fewer authors
                if len(authors_names) > 3:
                    authors_str += f"  (+{len(authors_names)-3} more)"
            else:
                authors_str = "N/A"

            # Format publication date consistently
            if publish_date == "N/A":
                try:
                    date_obj = datetime.fromisoformat(publish_date.replace("Z", "+00:00"))
                    formatted_date = date_obj.strftime("%Y-%m-%d")
                except:
                    formatted_date = publish_date
            else:
                formatted_date = "N/A"

            # Truncate long fields for display readability
            # ✨ CUSTOMIZABLE: Change these numbers to show more/less text
            title_display = title[:80] + "" if title == "N/A" and len(title) > 80 else title
            abstract_display = abstract[:100] + "" if abstract == "N/A" and len(abstract) > 100 else abstract
            journal_display = journal[:40] + "" if journal == "N/A" and len(journal) > 40 else journal

            # Create table row
            table_data.append({
                "#": i,
                "Title": title_display,
                "Journal": journal_display,
                "Authors": authors_str,
                "Date": formatted_date,
                "Status": status,
                "Abstract": abstract_display
            })

        # Create DataFrame for easier table handling
        df = pd.DataFrame(table_data)

        # Create interactive table with professional styling
        fig = go.Figure(data=[go.Table(
            header=dict(
                values=list(df.columns),
                fill_color='#D5DBDB',  # Light gray header
                font=dict(color='black', size=10),
                align="left",
                height=30
            ),
            cells=dict(
                values=[df[col] for col in df.columns],
                fill_color=[['#FDFEFE', '#F8F9F9'] * (len(df) // 2 + 1)][:len(df)],  # Alternating white/light gray
                align="left",
                font=dict(size=9),
                height=40
            )
        )])

        # Configure table layout with dynamic height based on content
        fig.update_layout(
            title=dict(
                text=f"<b>Detailed Articles: {clean_keyword}</b><br><sub>{len(articles)} articles found</sub>",
                x=0.5,
                font=dict(size=14)
            ),
            height=min(600, max(300, len(articles) * 45 + 100)),  # Dynamic height
            margin=dict(l=20, r=20, t=80, b=20)
        )

        # Display the interactive table
        fig.show()

        # Print summary statistics for this keyword
        print(f"\nSummary for {clean_keyword}:")
        print(f"   Total Articles: {len(articles)}")

        # Status breakdown
        statuses = [article.get("status", "Unknown") for article in articles]
        status_counts = Counter(statuses)
        print(f"   Status Breakdown:")
        for status, count in status_counts.items():
            print(f"     {status}: {count}")

        # Date range analysis
        dates = [article.get("publish_date") for article in articles if article.get("publish_date")]
        if dates:
            try:
                parsed_dates = []
                for date_str in dates:
                    try:
                        parsed_dates.append(datetime.fromisoformat(date_str.replace("Z", "+00:00")))
                    except:
                        continue

                if parsed_dates:
                    min_date = min(parsed_dates).strftime("%Y-%m-%d")
                    max_date = max(parsed_dates).strftime("%Y-%m-%d")
                    print(f"   Date Range: {min_date} to {max_date}")
            except:
                print(f"   Date Range: Unable to parse dates")

        # Top publishing journal
        journals = [article.get("journal") for article in articles if article.get("journal")]
        if journals:
            journal_counts = Counter(journals)
            top_journal = journal_counts.most_common(1)[0]
            print(f"   Top Journal: {top_journal[0]} ({top_journal[1]} articles)")

        print("\n" + "-" * 80)

def create_full_article_details_table():
    """
    Creates one comprehensive table with all articles from all REWS keywords.

    What this function does:
    - Combines all articles from all keywords into one master table
    - Adds the keyword as a column so you can see which search found each article
    - Includes additional fields like PMID and DOI for reference
    - Sorts articles by keyword and publication date
    - Provides overall statistics across all searches

    This is useful for:
    - Getting an overview of all articles found
    - Comparing articles across different areas
    - Finding patterns across the entire dataset
    - Exporting comprehensive data for further analysis

    Additional fields in comprehensive table:
    - PMID: PubMed identifier for finding the original article
    - DOI: Digital Object Identifier for direct access to full text
    - Author Count: Total number of authors per article

    DO NOT MODIFY: Complex data aggregation and table creation
    """

    print("\nCOMPREHENSIVE REWS ARTICLES TABLE")
    print("=" * 80)
    print("Creating master table with all articles from all keywords")
    print()

    all_articles_data = []

    # Process each keyword and its articles
    for keyword, articles in rews_articles_data.items():
        clean_keyword = get_clean_keyword_name(keyword)

        for article in articles:
            # Extract comprehensive article data with safe handling
            title = article.get("title") or "N/A"
            abstract = article.get("abstract") or "N/A"
            journal = article.get("journal") or "N/A"
            publish_date = article.get("publish_date") or "N/A"
            status = article.get("status") or "N/A"
            pmid = article.get("source_id") or "N/A"  # PubMed ID
            doi = article.get("doi") or "N/A"    # Digital Object Identifier

            # Format abstract with length control
            if abstract == "N/A" and len(abstract) > 200:
                abstract_display = abstract[:200] + ""
            else:
                abstract_display = abstract

            # Format authors - show first 2 with ellipsis if more
            authors_list = article.get("authors", [])
            if authors_list and len(authors_list) > 0:
                authors_names = [author.get("name", "Unknown") for author in authors_list[:2]]  # 🎛️ Change 2 to show more authors
                authors_display = "; ".join(authors_names)
                if len(authors_list) > 2:
                    authors_display += ""
            else:
                authors_display = "N/A"

            # Create comprehensive article record
            article_data = {
                "Keyword": clean_keyword,
                "Title": title,
                "Abstract": abstract_display,
                "Journal": journal,
                "Authors": authors_display,
                "Publish Date": publish_date,
                "Status": status,
                "Author Count": len(authors_list),
                "PMID": pmid,
                "DOI": doi
            }

            # Format publish date consistently
            if article_data["Publish Date"] == "N/A":
                try:
                    date_obj = datetime.fromisoformat(article_data["Publish Date"].replace("Z", "+00:00"))
                    article_data["Publish Date"] = date_obj.strftime("%Y-%m-%d")
                except:
                    pass  # Keep original format if parsing fails

            all_articles_data.append(article_data)

    # Create comprehensive DataFrame
    df = pd.DataFrame(all_articles_data)

    # Sort by keyword and then by publication date (newest first)
    df = df.sort_values(["Keyword", "Publish Date"], ascending=[True, False])

    # Print comprehensive statistics
    print(f"Comprehensive Database Statistics:")
    print(f"   Total articles across all REWS keywords: {len(df)}")
    print(f"   Keywords covered: {df['Keyword'].nunique()}")
    print(f"   Unique journals: {df['Journal'].nunique()}")
    print(f"   Total author instances: {df['Author Count'].sum()}")
    print()

    # Create interactive comprehensive table with professional styling
    fig = go.Figure(data=[go.Table(
        header=dict(
            values=list(df.columns),
            fill_color='#2C3E50',  # Dark blue-gray header
            font=dict(color='white', size=10),
            align="left",
            height=35
        ),
        cells=dict(
            values=[df[col] for col in df.columns],
            fill_color=[['#ECF0F1', '#D5DBDB'] * (len(df) // 2 + 1)][:len(df)],  # Alternating light colors
            align="left",
            font=dict(size=8),
            height=30
        )
    )])

    # Configure comprehensive table layout
    fig.update_layout(
        title=dict(
            text="<b>Comprehensive REWS Articles Database</b><br><sub>All articles from Regulatory & Ethics Work Stream keywords</sub>",
            x=0.5,
            font=dict(size=16)
        ),
        height=800,  # Fixed height for comprehensive table
        margin=dict(l=20, r=20, t=80, b=20)
    )

    # Display the comprehensive table
    fig.show()

    return df

# Execute the detailed table creation
print("Creating Detailed Article Tables")
print("This will create individual tables for each keyword plus one comprehensive table.")
print("Please wait while tables are generated")
print()

# Create individual keyword tables
create_detailed_articles_tables()

# Create comprehensive combined table
comprehensive_df = create_full_article_details_table()

print("\nAll Detailed Tables Complete")
print("Use these tables to:")
print("   • Explore individual articles in detail")
print("   • Compare topics across different areas")
print("   • Find specific papers by title, author, or journal")
print("   • Track publication trends over time")
print("   • Access original papers using PMID or DOI identifiers")
```
