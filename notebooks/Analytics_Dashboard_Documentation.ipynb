# Analytics Dashboard Documentation

This notebook walks through the APIs that feed the Analytics Dashboard and explains the JSON data they return from the database behind the dashboard. It includes clear examples of typical API requests and annotated JSON samples that show what each field means, plus practical guidance on querying, filtering, and interpreting the returned data. 

## AWS DB and Base URL

The tech team has curated data from the 3 sources for GitHub, Pypi and Pubmed and stored the data on AWS. The data within the DB can be accessed using APIs what we have provided.

The base URL is ```http://analytics-staging.ga4gh.org:8000```


## GitHub
1. ```/github/name/{repo-name}```: Fetch information for a single repository by name.
2. ```/github/owner/{owner-name}```: Fetch all repositories owned by the specified organization or user.
3. ```/github/all```: Fetch all repositories in the GitHub repos table.

### Json Payload
#### github_repos
1. ```id```: ID used by the DB
2. ```record_id```: ID used by the record table in the DB
3. ```name```: The name of the repo
4. ```repo_link```: The URL to the repo on GitHub
5. ```owner```: The owner organization or user 
6. ```description```: A short description of the repo within the repo
7. ```is_fork```: True if the repo is forked from another existing repo, false if not
8. ```last_updated```: Date when repository metadata (description, topics, visibility, etc.) was last changed.
9. ```pushed_at```: Date of last pushed commit
10. ```is_archived```: True if the repo is archived and no longer active, false if not
11. ```license```: License used by the repo
12. ```stargazers_count```: The number of stargazers for the repo
13. ```watchers_count```: The number of watchers for the repo
14. ```forks_count```: The number of forks for the repo
15. ```open_issues_count```: The number of open issues for the repo
16. ```network_count```: The amount of forks in the fork network (may include forks from parent repo)
17. ```subscribers_count```: The number of subscribers for the repo
18. ```branches_count```: The number of branches for the repo
19. ```created_by```: A field in the DB to track who the entry was added by
20. ```created_at```: A field in the DB to track when the entry was added
21. ```updated_by```: A field in the DB to track who the entry was updated by
22. ```updated_at```: A field in the DB to track when the entry was updated at
23. ```deleted_by```: A field in the DB to track who deleted the entry from the DB
24. ```deleted_at```: A field in the DB to track when the entry was deleted at
25. ```version```: A field in the DB to track the amount of times this entry was modified
26. ```created_on```: The date the repo was created on GitHub
